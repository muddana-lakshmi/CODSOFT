**1. Plot the count of males and females in the dataset.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
diwali_sales_df = pd.read_csv('/content/Diwali Sales Data.csv', encoding='latin1')
weather_history_df = pd.read_csv('/content/weatherHistory.csv', encoding='latin1')
print(diwali_sales_df.head())
gender_counts = diwali_sales_df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Count of Males and Females')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**2. Give the sum of amounts spent by each gender and plot the corresponding graph.**

```python
gender_col = 'Gender'
amount_spent_col = 'Amount'
gender_spending = diwali_sales_df.groupby(gender_col)[amount_spent_col].sum()
plt.figure(figsize=(8, 6))
gender_spending.plot(kind='bar',color='Red')
plt.title('Sum of Amounts Spent by Each Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount Spent')
plt.show()
```

**3. Count each age group and provide individual counts grouped by gender.**

```python
age_group_col = 'Age Group'
gender_col = 'Gender'
age_gender_counts = diwali_sales_df.groupby([age_group_col, gender_col]).size().unstack(fill_value=0)
print(age_gender_counts)
```

**4. Plot the total amount spent by each age group.**

```python
age_group_col = 'Age Group'
amount_spent_col = 'Amount'
age_group_spending = diwali_sales_df.groupby(age_group_col)[amount_spent_col].sum()
plt.figure(figsize=(10, 6))
age_group_spending.plot(kind='bar',color='cyan')
plt.title('Total Amount Spent by Each Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Amount Spent')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**5. Plot a graph depicting the total number of orders from the top 10 states.**

```python
state_col = 'State'
orders_col = 'Orders'
state_orders = diwali_sales_df.groupby(state_col)[orders_col].sum()
top_10_states = state_orders.nlargest(10)
plt.figure(figsize=(12, 8))
top_10_states.plot(kind='bar',color='Green')
plt.title('Total Number of Orders from the Top 10 States')
plt.xlabel('State')
plt.ylabel('Total Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**6. Determine the total amount spent in the top 10 states.**

```python
print(diwali_sales_df.head())
state_col = 'State'
amount_spent_col = 'Amount'
state_amount_spent = diwali_sales_df.groupby(state_col)[amount_spent_col].sum()
top_10_states_amount = state_amount_spent.nlargest(10)
print("Total Amount Spent in the Top 10 States:")
print(top_10_states_amount)
plt.figure(figsize=(12, 8))
top_10_states_amount.plot(kind='bar',color='Yellow')
plt.title('Total Amount Spent in the Top 10 States')
plt.xlabel('State')
plt.ylabel('Total Amount Spent')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**7. Plot a comparison graph between the number of married and unmarried individuals.**

```python
print(diwali_sales_df.head())
marital_status_col = 'Marital_Status'
marital_status_counts = diwali_sales_df[marital_status_col].value_counts()
plt.figure(figsize=(8, 6))
marital_status_counts.plot(kind='bar', color=['red', 'green'])
plt.title('Comparison of Married vs Unmarried Individuals')
plt.xlabel('Marital Status')
plt.ylabel('Number of Individuals')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**8. Plot the amount spent by males and females based on marital status.**

```python
gender_col = 'Gender'
marital_status_col = 'Marital_Status'
amount_spent_col = 'Amount'
gender_marital_spending = diwali_sales_df.groupby([gender_col, marital_status_col])[amount_spent_col].sum()
gender_marital_spending = gender_marital_spending.unstack(level=1)
plt.figure(figsize=(10, 6))
gender_marital_spending.plot(kind='bar', width=0.8)
plt.title('Amount Spent by Gender and Marital Status')
plt.xlabel('Gender')
plt.ylabel('Total Amount Spent')
plt.xticks(rotation=0)
plt.legend(title='Marital Status')
plt.tight_layout()
plt.show()
```

**9. Plot the count of each occupation present in the dataset.**

```python
occupation_col = 'Occupation'
occupation_counts = diwali_sales_df[occupation_col].value_counts()
plt.figure(figsize=(10, 6))
occupation_counts.plot(kind='bar',color='gray')
plt.title('Count of Each Occupation')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**10. Plot the amount spent by each occupation in descending order.**

```python
occupation_col = 'Occupation'
amount_spent_col = 'Amount'
occupation_amount_spent = diwali_sales_df.groupby(occupation_col)[amount_spent_col].sum()
occupation_amount_spent = occupation_amount_spent.sort_values(ascending=False)
plt.figure(figsize=(12, 8))
occupation_amount_spent.plot(kind='bar')
plt.title('Amount Spent by Each Occupation (Descending Order)')
plt.xlabel('Occupation')
plt.ylabel('Total Amount Spent')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**11. Provide statistical analysis of each product category based on the percentage of orders completed.**

```python
import pandas as pd
diwali_sales_data = pd.read_csv('/content/Diwali Sales Data.csv', encoding='ISO-8859-1')
print(diwali_sales_data.columns)
product_category_column = 'Product_Category'
orders_column = 'Orders'
status_column = 'Status'
diwali_sales_data['Completed_Order'] = diwali_sales_data[status_column].apply(lambda x: 1 if x == 'Completed' else 0)
category_stats = diwali_sales_data.groupby(product_category_column).apply(
lambda x: (x['Completed_Order'].sum() / x[orders_column].sum()) * 100).reset_index(name='Order_Completed_Percentage')
category_stats_desc = diwali_sales_data.groupby(product_category_column)['Completed_Order'].describe()
print("Statistical Analysis of Each Product Category Based on Percentage of Orders Completed:")
print(category_stats_desc)
print("\nOrder Completion Percentage by Product Category:")
print(category_stats)
```

**12. Determine the budget spent on each product category in descending order.**

```python
data = pd.read_csv('/content/Diwali Sales Data.csv', encoding='unicode_escape')
Budget_spent = data.groupby('Product_Category').agg(total_amount = ('Amount', 'sum')).reset_index()
Budget_spent = Budget_spent.sort_values(by = 'total_amount', ascending=False)
print(Budget_spent)
```
